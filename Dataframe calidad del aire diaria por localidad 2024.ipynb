# Dataframe calidad del aire diaria por localidad 2024

```python
import numpy as np
import pandas as pd

ds1 = 'Calidad aire 2024.csv'
dfair = pd.read_csv(ds1, sep=";", header=[0, 1], low_memory=False)
dfair2= dfair.copy()
dfair2.head()
```

```python
dfair2['Unnamed: 0_level_0', 'Fecha & Hora'] = pd.to_datetime(dfair2['Unnamed: 0_level_0', 'Fecha & Hora'], dayfirst=True, errors="coerce")
```

```python
dfair2.describe(include='all')
```

```python
col_fecha = ('Unnamed: 0_level_0', 'Fecha & Hora')
dfair2["fecha"] = dfair2[col_fecha].dt.date

# Para asegurar que los datos sean numéricos y consistentes
def to_num(s):
    return pd.to_numeric(s.astype(str).str.replace("Sin data", "", case=False, regex=False).str.replace(",", ".", regex=False).str.strip(),errors="coerce")

def buscar_columna(sub, texto1, texto2=None):
    for c in sub.columns:
        nombre = str(c).lower()
        if texto1.lower() in nombre and (texto2 is None or texto2.lower() in nombre):
            return c
    return None

tidy_list = []

# Lista de nombres de estaciones
estaciones = [
    est for est in dfair2.columns.get_level_values(0).unique()
    if est not in [col_fecha[0], "fecha"]
]

# DataFrame con las columnas de esa estación
for est in estaciones:
    sub = dfair2.xs(est, axis=1, level=0).copy()   

    # PM2.5
    c25_c = buscar_columna(sub, "pm2.5", "concentr")   # concentración PM2.5
    c25_n = buscar_columna(sub, "pm2.5", "nowcast")    # nowcast PM2.5
    c25_i = buscar_columna(sub, "pm2.5", "iboca")      # IBOCA PM2.5

    # PM10
    c10_c = buscar_columna(sub, "pm10", "concentr")
    c10_n = buscar_columna(sub, "pm10", "nowcast")
    c10_i = buscar_columna(sub, "pm10", "iboca")

    df_est = pd.DataFrame({
        "fecha & hora": dfair2[col_fecha],
        "fecha": dfair2["fecha"],
        "estación": est,
        "pm25_conc": to_num(sub[c25_c]) if c25_c is not None else np.nan,
        "pm25_nowcast": to_num(sub[c25_n]) if c25_n is not None else np.nan,
        "pm25_iboca": to_num(sub[c25_i]) if c25_i is not None else np.nan,
        "pm10_conc": to_num(sub[c10_c]) if c10_c is not None else np.nan,
        "pm10_nowcast": to_num(sub[c10_n]) if c10_n is not None else np.nan,
        "pm10_iboca": to_num(sub[c10_i]) if c10_i is not None else np.nan,
    })

    tidy_list.append(df_est)

tidy = (pd.concat(tidy_list, ignore_index=True).sort_values(["estación", "fecha & hora"]).reset_index(drop=True))

# CSV de estaciones y su respectiva localidad
est = pd.read_csv("estaciones_calidad_aire.csv", sep=";")

# Unir localidad al dataframe tidy
tidy = tidy.merge(est[["estación", "localidad"]], on="estación", how="left")

def agg_loc_day(g):
    cols = ["pm25_conc", "pm25_nowcast", "pm25_iboca","pm10_conc", "pm10_nowcast", "pm10_iboca"]
    res = {}
    for col in cols:
        res[col] = g[col].mean() 
    return pd.Series(res)
    
df_resumen = (tidy[tidy["localidad"].notna()].groupby(["fecha", "localidad"]).apply(agg_loc_day, include_groups=False)  .reset_index().sort_values(["localidad","fecha"]).reset_index(drop=True))
```

```python
# Ahora, para imputar los datos faltantes
df_resumen["fecha"] = pd.to_datetime(df_resumen["fecha"], dayfirst=True, errors="coerce")
cols = ["pm25_conc", "pm25_nowcast", "pm25_iboca","pm10_conc", "pm10_nowcast", "pm10_iboca"]
# Para la imputación, esta se realiza con el promedio del mes en esa localidad. En caso de que no hayan datos en todo el mes, se imputa con el promedio anual de la variable en esa localidad
for col in cols:
    # media por localidad-año-mes (por métrica)
    media_mes = (df_resumen.groupby(["localidad", df_resumen["fecha"].dt.year, df_resumen["fecha"].dt.month])[col].transform("mean"))

    # media por localidad-año (por métrica)
    media_año = (df_resumen.groupby(["localidad", df_resumen["fecha"].dt.year])[col].transform("mean"))

    # imputar: primero con media del mes, si sigue NaN usar media del año
    df_resumen[col] = df_resumen[col].fillna(media_mes).fillna(media_año)

# Para añadir columna de zona según localidad 
zonas = pd.read_csv("Localidades Bogotá por zona.csv")

# Aplico llave normalizada para evitar problemas de mayúsculas/acentos
zonas["loc_key"] = zonas["Localidad"].str.upper().str.strip()
df_resumen["loc_key"] = df_resumen["localidad"].str.upper().str.strip()

# Merge para traer la zona
df_resumen = df_resumen.merge(zonas[["loc_key", "Zona"]],on="loc_key",how="left")

# Limpieza y nombre final de la columna
df_resumen = df_resumen.drop(columns=["loc_key"])
df_resumen = df_resumen.rename(columns={"Zona": "zona"})

cols = list(df_resumen.columns)
idx_loc = cols.index("localidad")
# Insertar 'zona' justo después de 'localidad'
cols = cols[:idx_loc + 1] + ["zona"] + cols[idx_loc + 1:]

df_resumen = df_resumen[cols]

# Guardar versión imputada
df_resumen.to_csv("resumen_localidad_diario_pm25_pm10_imp.csv", index=False)
```
