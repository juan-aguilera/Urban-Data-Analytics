```python
import pandas as pd

#Cargar archivo "Base Anuario de Siniestralidad 2024.xlsx" 

df = pd.read_excel("Base_Anuario_de_Siniestralidad_2024.xlsx",sheet_name = "Siniestros")
print(df.head(5))
```

```python
print(df.columns)
```

```python
#Eliminar columnas que no son necesarias 

cols_eliminar = ["Codigo_Accidente","Formulario","Direccion","MM_Acc","DD_Mes_Acc","Dia_Semana_Acc"
                ,'Min_Acc','Clase_Acc', 'Elemento_Choque','Tipo_Objeto_Fijo','Gravedad_indicador_30d', 'Con_Bicicleta',
                 'Con_Carga','Con_Huecos', 'Con_Menores', 'Con_Moto', 'Con_Peaton','Con_Persona_Mayor', 'Con_Rutas', 'Con_Tpi', 'Con_Tpp', 'Con_Velocidad',
                'Con_Sitp', 'Con_Troncal', 'Con_Alimentador', 'Con_Zonal','Con_Provisional', 'Con_Articulado', 'Con_Biarticulado',
                'Con_Padron_Dual', 'Con_Servicio_Especial', 'Con_Taxi']
df_nuevo = df.drop(columns = cols_eliminar)
print(df_nuevo.head(5))
```

```python
print(df_nuevo.dtypes)
```

```python
#Eliminar todos los registros que no pertenecen al año 2024 
df_2024 = df_nuevo[df_nuevo["AA_Acc"] == 2024]
print(df_2024.shape)
```

#¿Hay valores NA?
df_2024.isna().sum()

```python
#Remplazar valores 
df_2024["Con_Embriaguez"] = df_2024["Con_Embriaguez"].fillna("NO")
df_2024.isna().sum()
```

```python
#Conocer localidades que tiene el dataset
localidades = sorted(df["Localidad"].unique())
print(localidades)
```

Coinciden con las que tiene la tabla **"Validaciones_localidad_mensual_2024"**

```python
#Asignar zona a cada localidad 
loc_zona = pd.read_excel("Preguntas Bogota como vamos.xlsx",sheet_name ="Localidades-zonas")
loc_zona["Localidad"]= loc_zona["Localidad"].str.upper()
loc_zona["Localidad"] = loc_zona["Localidad"].replace("LA CANDELARIA","CANDELARIA")
print(loc_zona)
```

```python
#Hacer un left join para crear una nueva columna "Zona" y asignarle una zona a cada localidad 
df_2024 = pd.merge(df_2024, loc_zona, on="Localidad", how="left")
print(df_2024.head(50))
```

```python
df_2024.to_csv("Siniestralidad_2024.csv",sep = ",", index = False, encoding = "utf-8-sig")
```
