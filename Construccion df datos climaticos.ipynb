```python
from unidecode import unidecode
```

```python
#Combinar las diferentes versiones de .csv generadas

import pandas as pd
import glob

files = glob.glob("datos climaticos por localidad/*.csv")


```

```python
dataframes = [pd.read_csv(f) for f in files]
df_total = pd.concat(dataframes, ignore_index = True)
```

```python
print(df_total.head(20))
print(df_total.shape)
```

```python
#Eliminar filas duplicadas
df_total = df_total.drop_duplicates(keep='first')
```

```python
print(df_total.head(20))
print(df_total.shape)
```

```python
#Verificar cantidad de registros por localidad
df_total["Localidad"].value_counts()
```

```python
#Agregar zona a cada localidad 
loc_zona = pd.read_csv("Localidades Bogotá por zona.csv")
#Normalizar columna
loc_zona['Localidad'] = loc_zona['Localidad'].apply(lambda x: unidecode(str(x)).strip())
loc_zona['Localidad'] = loc_zona['Localidad'].replace('La Candelaria','Candelaria')
print(loc_zona)
```

```python
df_con_zonas = pd.merge(df_total,loc_zona, on="Localidad",how='left')
df_con_zonas['Localidad'] = df_con_zonas['Localidad'].apply(lambda x: unidecode(str(x).upper())) 
print(df_con_zonas.head(10))
```

```python
df_con_zonas.to_csv("datos climaticos por localidad.csv",sep=',',index=False,encoding='utf-8-sig')
```
